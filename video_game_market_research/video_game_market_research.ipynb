# Анализ рынка компьютерных игр

## Описание проекта:

В данном проекте исследуется рынок компьютерных игр за период **1980-2016 гг.**, и рассматриваются актуальные тренды для последующего выявления наиболее прибыльных сегментов рынка и эффективного планирования рекламных кампаний. Исследование рынка затронуло анализ релизов, оценок и глобальных продаж игр, и по его результатам были составлены портреты типичных покупателей в различных географических регионах. В ходе работы были получены интересные инсайты по различным платформам, жанрам и потребителям в целом, а также в разрезе конкретных регионов.

## Главные выводы:

* Начиная с 2002 г., общий рост числа релизов во многом поддерживался за счет перевыпуска игр на различные платформы, которые со временем стали более многочисленными.


* В большинстве случаев с начала продаж игр на платформу до пика проходит около 5 лет. Компании фундаментально обновляют свои линейки примерно каждые 6 лет. В среднем платформы остаются актуальными 6-7 лет. Консоли нового поколения вышли в 2013 г. (WiiU - в 2012 г.) Таким образом, следующая смена поколения должна наступить в 2019-2020 гг.


* В общем по всем платформам связь между оценками критиков и продажами умеренная и положительная. На уровне различных платформ сила связи может меняться, но чем платформа новее, тем более чувствительны продажи игр на нее к отзывам критиков, то есть чем выше оценки - тем выше продажи игр. Связь между оценками пользователей и продажами отсутствует в общем, однако на уровне конкретных платформ выраженная положительная связь была найдена только для 3DS и WiiU. 


* За исследуемый период больше всего, со значительным отрывом, было выпущено игр жанра Action (752 релиза или 34.8% от общего числа). Иными словами, на Action приходится более трети всех релизов. На втором месте Role-Playing (RPG) (285 релизов или 13.2%). На третьем месте Adventure (211 релизов или 9.8%). Также большое количество игр было выпущено в жанрах Sports (208 релизов или 9.6%) и Shooter (187 релизов или 8.7%). 


* По глобальным продажам жанр Action снова является бесспорным лидером (29.5% всех продаж). Однако далее топ жанров по выпуску и топ по продажам имеют различия. На втором месте по продажам расположился жанр Shooter (21.5% всех продаж). На третьем месте Sports (13.8%). Ролевые игры на четвертом месте (13.3%). Продажи всех остальных жанров в несколько раз ниже. Топ-4 жанров по продажам занимают почти 78% всего рынка. На остальные 8 жанров приходится лишь 22% всех продаж.


* В Северной Америке лидирует PS4, за ней следует с небольшим отставанием XOne. В Европе также очевидно доминирует Sony PlayStation. При этом американский рынок более диверсифицирован - консоли Sony и Microsoft балансируют друг друга. Япония же - рынок только для "отечественных" консолей. Поскольку Япония является родиной сразу двух производителей приставок - Sony и Nintendo, платформы этих компаний безоговорочно контролируют весь внутренний рынок. Кроме того, Япония - единственный регион, где самой популярной платформой является портативная консоль - а именно 3DS. 


* Доля Северной Америки в глобальном рынке игр - 40.27%, Европы - 36.03%, Японии - 12.66%. 


* Продажи зависят от ESRB рейтинга игр: в Северной Америке и Европе наблюдается некоторая схожесть - наиболее популярны игры категории Mature. Япония совсем не похожа на другие регионы - самой популярной ESRB категорией является Teen.


* По результатам статистических тестов, не удалось отклонить гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC равны. Таким образом, наше предположение о том, что средние рейтинги на этих двух платформах одинаковые, можно признать обоснованным.


* Удалось выявить статистически значимое различие между средними пользовательскими рейтингами жанров Action и Sports.

* Рассмотрим проблемы, связанные с преобразованием типов данных и пропущенными значениями.

`year_of_release` рационально привести к целочисленному типу, так как год - это целое число. Конвертация в формат datetime избыточна, поскольку у нас имеются данные лишь по годам. Однако в данном столбце присутствуют несуществующие значения, что делает изменение типа на целочисленный невозможным. Рассмотрим, насколько строки с пропущенными годами являются релевантными для анализа. Посмотрим на платформы, для которых были выпущены те игры, чтобы примерно определить временной промежуток.

```python
data[data['year_of_release'].isna()]['platform'].value_counts()
```

Видим, что значительную часть платформ в данном срезе можно отнести к устаревшим или близким к устареванию (на 2016 г.) и, соответственно, нерелевантным в рамках нашего анализа. Таким образом, мы можем устранить все строки с пропущенными значениями года выпуска (т.к. доля этих строк незначительна, а заполнение затруднительно) и далее привести столбец к целочисленному типу. Возможно, данные пропуски возникли вследствие ошибки экспорта данных.

```python
data.dropna(subset=['year_of_release'], inplace=True)
data['year_of_release'] = data['year_of_release'].astype('int')
```

Также устраним пропуски в столбцах `genre` и `name`, поскольку данная информация не подлежит заполнению, а количество пропущенных значений пренебрежимо мало. Ошибки также, скорее всего, возникли вследствие неправильной выгрузки данных.

```python
data.dropna(subset=['genre', 'name'], inplace=True)
data.info()
```

Далее взглянем, почему тип `user_score` определился как object вопреки ожиданиям. Возможно в столбец попали некорректные значения.

```python
data['user_score'].unique()
```

```python
data.query('user_score == "tbd"').shape[0] / data.shape[0]
```

Видим, что в столбце `user_score` оказалось значение 'tbd', которое повлияло на определение типа. 'tbd' значит 'to be determined' - это говорит о том, что пользователь еще не успел поставить оценку игре. Таких строк 14.45% в датафрейме. По смыслу в данном контексте 'tbd' эквивалентно пропущенному значению. Мы не можем заменить его нулем, поскольку 0 - это худшая оценка, а пользователи просто не проголосовали. Таким образом, рациональнее всего заменить 'tbd' на null значения методом `to_numeric`.

```python
data['user_score'] = pd.to_numeric(data['user_score'], errors='coerce')
data.info()
```

```python
data.isna().mean().sort_values(ascending=False)
```

После первичной обработки, пропуски остались в 3 столбцах: `user_score`, `critic_score`, `rating`

Более половины значений в `user_score` и `critic_score` пропущены, и заполнять их совершенно некорректно и не имеет практического смысла. Дело в том, что при половине пропусков, заполнение средним, медианой, или модой (как общими показателями, так и по группам) крайне сильно исказит распределение и снизит репрезентативность данных. Поэтому на данном этапе было решено оставить все строки для анализа, но для более детального рассмотрения оценок пользователей и критиков при необходимости можно создать дополнительные датафреймы с исключенными пропусками.

### **3.1. Рассмотрим какое количество игр выпускалось в разные годы** <a name="subparagraph3_1"></a>

* Проанализируем общее количество выпускаемых игр по всем платформам и сохраним сгруппированные данные в датафрейме `games_by_years`. Также посмотрим на количество уникальных тайтлов - для этого оставим только первые записи об играх в датафрейме `unique_games_by_years`, а последующие релизы на другие платформы исключим.

```python
games_by_years = data.groupby('year_of_release')['name'].agg('count')
unique_games = data.drop_duplicates(subset=['name'], keep='first')
unique_games_by_years = unique_games.groupby('year_of_release')['name'].agg('count')
pivot_by_years = pd.concat([games_by_years, unique_games_by_years], axis=1)
pivot_by_years.columns = ['number_of_games_total', 'number_of_unique_titles']
pivot_by_years.transpose()
```

```python
ax = games_by_years.plot(kind='bar', rot='0', alpha=0.5, label = 'Общее количество релизов', legend=True)
unique_games_by_years.plot(kind='bar', ax=ax, rot='25', label = 'Количество уникальных игр', legend=True)
plt.title('Количество выпущенных игр по годам')
plt.xlabel('')
plt.ylabel('Количество игр')
plt.gcf().set_size_inches(16, 8)
plt.legend(loc='upper left')
plt.show()
```

**Вывод:** 

На графике выше можно увидеть, что в 80-е годы игровая индустрия была нишевой и находилась в относительной стагнации - количество выпускаемых игр было невелико. С начала 90-х наблюдается развитие отрасли, и количество игр начинает быстро увеличиваться. С 2000 года рост числа игр еще сильнее ускоряется и достигает пика в 2008 году. Экономический кризис 2008 г., по-видимому, оказал существенное влияние на игровые компании, что привело к последующему ежегодному снижению числа релизов. *Стабилизация отрасли наступила лишь в период 2013-2016 гг.*

Важно отметить, что, начиная с 2002 г., общий рост числа релизов во многом поддерживался за счет перевыпуска игр на различные платформы, которые со временем стали более многочисленными. Таким образом, можно считать, что "бум" кросс-платформенности начался в 2002 г. До этого, перевыпуски занимали крайне небольшую долю от общего количества релизов в году, и в основном каждый год выпускались новые игры. После 2002 г. повторные релизы на других платформах стали занимать существенную долю от общего числа выпусков игр. Также любопытно, что по количеству уникальных игр период 2013-2016 гг. соответствует второй половине 90-х годов, однако за счет повторных релизов на другие платформы общее число игр значительно выше, чем в тот период.

На основании имеющихся данных можно предположить, что самым репрезентативным в рамках нашего анализа является *период относительной стабильности: 2013-2016 гг.* Предшествующие периоды полезны для оценки общего движения отрасли, однако, они не подходят для краткосрочного прогнозирования вследствие быстро меняющихся тенденций в отрасли.

* Обобщим все имеющиеся данные по жизненному циклу различных платформ. Предварительно соберем необходимые данные по началу и окончанию продаж игр в датафрейм `lifespan_chart_data` для последующей визуализации. Платформы, для которых отсутствуют продажи на 2016 г. будем считать неактуальными.

```python
test_platform_lifespan_raw = platform_lifespan_raw.reset_index()
test_platform_lifespan_raw.columns = test_platform_lifespan_raw.columns.rename('platform')

supertest_platform_lifespan_raw = test_platform_lifespan_raw.transpose().drop(['platform'],axis=0)
first_valid_indices = supertest_platform_lifespan_raw.apply(lambda series: series.first_valid_index())
last_valid_indices = supertest_platform_lifespan_raw.apply(lambda series: series.last_valid_index())

lifespan_chart_data = test_platform_lifespan_raw[['platform']].copy()
lifespan_chart_data.columns = lifespan_chart_data.columns.rename('index')
lifespan_chart_data['start_date'] = first_valid_indices
lifespan_chart_data['end_date'] = last_valid_indices
lifespan_chart_data = lifespan_chart_data.sort_values(by='start_date', ascending=False).reset_index(drop=True)
lifespan_chart_data['start_date'] = pd.to_datetime(lifespan_chart_data['start_date'], format='%Y')
lifespan_chart_data['end_date'] = pd.to_datetime(lifespan_chart_data['end_date'], format='%Y')
lifespan_chart_data['is_relevant'] = np.where(lifespan_chart_data['end_date'] < '2016', False, True)
```

* Далее построим общий таймлайн и нанесем на него продолжительность жизни различных платформ.

```python
years = mdates.YearLocator()
date_fmt = mdates.DateFormatter('%Y')

lifespan_chart_data['duration'] = lifespan_chart_data['end_date'] - lifespan_chart_data['start_date']

tasks = {task:(i+1)*10 for i, task in enumerate(dict.fromkeys(lifespan_chart_data['platform']))}

start_times = [np.datetime64(t, 'Y') for t in lifespan_chart_data['start_date']]

fig, ax_gnt = plt.subplots(figsize = (16, 8))

for i, t in enumerate(lifespan_chart_data['platform']):
    ax_gnt.broken_barh([(start_times[i], lifespan_chart_data['duration'][i])], (tasks[t], 9), facecolors = ('tab:cyan') if lifespan_chart_data['is_relevant'][i] else ('tab:grey')) 

ax_gnt.set_yticks([i + 5 for i in tasks.values()]) 
ax_gnt.set_yticklabels(dict.fromkeys(lifespan_chart_data['platform']).keys()) 

ax_gnt.xaxis_date()
ax_gnt.xaxis.set_major_locator(years)
ax_gnt.xaxis.set_major_formatter(date_fmt)
fig.autofmt_xdate()

plt.title('Жизненный цикл платформ за период 1980 - 2016 (цветные - актуальные) \n')

plt.show()
```

**Вывод:**

На данном графике хорошо отображается одновременное нахождение на рынке различных платформ в разные периоды времени, а также видно, что исторически новые платформы от различных компаний появлялись каждые 2-3 года, однако с 2000 года расстояние между анонсами новых платформ на рынке несколько увеличилось. Это может быть связано с тем, что и платформы стали более технологичными и сложными, а значит они требуют больше R&D перед выпуском на рынок.

Как мы выяснили ранее на основе исторических данных, платформы в среднем остаются актуальными в течение 6-7 лет с выпуска, однако продажи игр начинают падать по мере устаревания консоли и выхода нового поколения.

Ранее мы отметили, что продажи игр для предыдущего поколения стремительно снижаются - консоли уже долго находятся на рынке (более 7 лет, значит их жизненный цикл подходит к завершению). В случае корреляции неправомерно говорить о причинно-следственной зависимости, поэтому мы можем только сказать, что довольно часто игры, которые были положительно оценены критиками, также были положительно оценены пользователями и наоборот. Нельзя утверждать, что оценки критиков влияют на пользователей или что оценки пользователей влияют на критиков. 

Мы можем резюмировать, что игроки чаще прислушиваются к мнению критиков при покупке игры, а не к мнению друг друга, но в итоге довольно часто оценивают игры похожим образом.

### 3.7. Связь между отзывами на игры и продажами на различных платформах <a name="subparagraph3_7"></a>

Построим диаграммы рассеяния и рассчитаем коэффициенты корреляции для каждой платформы. И далее сравним, есть ли отличия на самой популярной PS4 от других платформ.

* Связь между оценкой критиков и продажами игр на разных платформах

```python
# Для Wii (под индексом 6) недостаточно данных - удаляем из массива с названиями платформ.
# Пишем цикл для всех остальных платформ.
for i in np.delete(games['platform'].unique(), 6):
    platform = games[games['platform'] == i].copy()
# Строим скаттерплоты для каждой платформы.
    platform.plot(kind='scatter', x='critic_score', y='total_sales', alpha=0.8)
    plt.title(f'Диаграмма рассеивания между оценкой критиков и продажами игр для {i}')
    plt.xlabel('\n Оценка критиков по 100-балльной шкале')
    plt.ylabel('Продажи игр в млн. копий')
    plt.gcf().set_size_inches(16, 6)
    plt.show()
# Считаем корреляцию Пирсона между оценками критиков и общими продажами.
    corr_critic_sales = platform['critic_score'].corr(platform['total_sales'])
    print(f'Коэффициент корреляции между отзывами критиков и продажами для {i}: {corr_critic_sales:.2f}')
```

Чуть слабее данная связь, но все равно существенная, для `WiiU` (**0.38**), `3DS` (**0.36**), `X360` (**0.35**) и `PS3` (**0.33**).

Можем обобщить, что чем платформа новее, тем более чувствительны продажи игр на нее к отзывам критиков: чем выше оценки - тем выше продажи игр. В случае с `PC` доля пиратства по-прежнему высока, и значительная доля игроков может самостоятельно скачать нелицензионную копию игры и ознакомиться с ней, поэтому роль критиков самая слабая.

## 5. Проверка статистических гипотез <a name="paragraph5"></a>

* Проверим гипотезу о том, что средние пользовательские рейтинги платформ `Xbox One` и `PC` одинаковые.

Сформулируем нулевую и альтернативную гипотезы: H0 и H1.

    H0: средние пользовательские рейтинги платформ Xbox One и PC равны
    H1: средние пользовательские рейтинги платформ Xbox One и PC не равны

Сохраним выборки в `xone_sample` и `pc_sample`

```python
xone_sample = games.query('platform == "XOne"').dropna(subset=['user_score'])['user_score']
pc_sample = games.query('platform == "PC"').dropna(subset=['user_score'])['user_score']
```

Проверим выполняется ли требование гомогенности дисперсии с помощью критерия Левена

    H0: выборки взяты из генеральных совокупностей с равными дисперсиями
    H1: выборки взяты из генеральных совокупностей с различными дисперсиями

```python
alpha = 0.05
results = st.levene(xone_sample, pc_sample)
print('p-value:', results.pvalue)
if results.pvalue < alpha:
    print('Нулевая гипотеза о равенстве дисперсий отвергается. Дисперсии гетерогенны.')
else:
    print('Нет оснований отвергать нулевую гипотезу о равенстве дисперсий. Дисперсии гомогенны.')
```

Так как требование гомогенности дисперсии не выполняется, будем использовать параметр `equal_var=False` для парного t-теста.

```python
alpha = 0.05
results = st.ttest_ind(xone_sample, pc_sample, equal_var=False)
print('p-value:', results.pvalue)
if results.pvalue < alpha:
    print('Нулевая гипотеза о равенстве средних отвергается. Принимается альтернативная гипотеза о том, что средние различаются.')
else:
    print('Нет оснований отвергать нулевую гипотезу о равенстве средних.')
```

**Вывод:** согласно результатам теста, у нас нет оснований отклонить нулевую гипотезу о том, что средние пользовательские рейтинги платформ `Xbox One` и `PC` равны. Таким образом, наше предположение о том, что средние рейтинги на этих двух платформах одинаковые, можно признать обоснованным.

* Проверим гипотезу о том, что средние пользовательские рейтинги жанров `Action` и `Sports` различаются

Сформулируем нулевую и альтернативную гипотезы: H0 и H1.

    H0: средние пользовательские рейтинги жанров `Action` и `Sports` равны
    H1: средние пользовательские рейтинги жанров `Action` и `Sports` не равны

Сохраним выборки в `action_sample` и `sports_sample`

```python
action_sample = games.query('genre == "Action"').dropna(subset=['user_score'])['user_score']
sports_sample = games.query('genre == "Sports"').dropna(subset=['user_score'])['user_score']
```

Проверим выполняется ли требование гомогенности дисперсии с помощью критерия Левена

    H0: выборки взяты из генеральных совокупностей с равными дисперсиями
    H1: выборки взяты из генеральных совокупностей с различными дисперсиями

```python
alpha = 0.05
results = st.levene(action_sample, sports_sample)
print('p-value:', results.pvalue)
if results.pvalue < alpha:
    print('Нулевая гипотеза о равенстве дисперсий отвергается. Дисперсии гетерогенны.')
else:
    print('Нет оснований отвергать нулевую гипотезу о равенстве дисперсий. Дисперсии гомогенны.')
```

Так как требование гомогенности дисперсии не выполняется, будем использовать параметр `equal_var=False` для парного t-теста.

```python
alpha = 0.05
results = st.ttest_ind(action_sample, sports_sample, equal_var=False)
print('p-value:', results.pvalue)
if results.pvalue < alpha:
    print('Нулевая гипотеза о равенстве средних отвергается. Принимается альтернативная гипотеза о том, что средние различаются.')
else:
    print('Нет оснований отвергать нулевую гипотезу о равенстве средних.')
```

**Вывод:** согласно результатам теста, мы можем отвергнуть нулевую гипотезу о том, что средние пользовательские рейтинги жанров `Action` и `Sports` равны. Таким образом, вполне обосновано наше предположение о том, что *средние пользовательские рейтинги этих двух жанров различаются*, и мы можем утверждать, что *данное различие является статистически значимым*.

* **В Северной Америке** лидирует PS4, за ней следует с небольшим отставанием XOne. После лидеров нового поколения идут PS3 и X360. Значительно меньшей популярностью по сравнению с лидерами пользуются 3DS и WiiU. Аутсайдерами в регионе можно считать PC, Wii, PSV. Можно резюмировать, что **американский рынок более диверсифицирован** - консоли Sony и Microsoft балансируют друг друга. **В Европе** очевидно **доминирует** влияние **Sony PlayStation** (PS4 (продаж больше, чем в Сев.Америке) и PS3). Но консоли Microsoft следуют за лидерами и тоже находятся в топе по продажам игр (XOne и X360). Умеренной популярностью также пользуются 3DS, PC (продаж больше, чем в других регионах) и WiiU. Аутсайдеры: PSV, Wii. **Япония** - рынок только для "отечественных" консолей. Поскольку Япония является родиной сразу двух производителей приставок - Sony и Nintendo, платформы этих компаний безоговорочно контролируют весь внутренний рынок. Для приставок Microsoft нет места - их вероятно покупают только энтузиасты. Игры на PC также не интересуют японцев. Отличительной особенностью является то, что **Япония - единственный регион, где самой популярной платформой является портативная консоль** - а именно 3DS, причем с существенным отрывом. 


* **Доля Северной Америки** в глобальном рынке игр: **40.27%**, **Европы** - **36.03%**, **Японии** - **12.66%**.


* **В Европе и Северной Америке топ-5 популярных жанров** практически идентичен за исключением пятого места:
    1. Action
    2. Shooter
    3. Sports
    4. Role-Playing
    5. Racing - в Европе / Misc. - в Америке (игры, не вписывающиеся в определенную категорию)


* **В Японии** в силу культурных различий с остальным миром, **топ-5 популярных жанров** выглядит иначе (На самом деле, японцы в подавляющем большинстве случаев предпочитают лишь 2 жанра - ролевые игры и экшн. (даже третье место отстает почти в 4 раза по продажам от второго):
    1. Role-Playing
    2. Action
    3. Misc (в случае Японии, в данную категорию, скорее всего, могли попасть игры с аниме-тематикой, крайне популярной в стране)
    4. Fighting
    5. Shooter
  
  
* **Продажи также зависят от ESRB рейтинга игр**: **в Северной Америке и Европе** снова наблюдается некоторая схожесть - **наиболее популярны** с большим отрывом игры категории **Mature**. **Вторая по популярности** - категория **Everyone**. Популярность игр категории E10+ и Teen примерно одинакова.


* **Япония** совсем не похожа на другие регионы: **самой популярной ESRB категорией** является **Teen**. Продажи игр категории Everyone даже немного превысили Mature. **Наименее популярная категория - E10+**, в ней продажи существенно ниже, чем в других категориях. Можем заключить, что **японцы нейтрально относятся к "взрослым" играм** с жестокостью и кровью, **в отличие от европейцев и американцев**.


* По результатам стат. тестов, у нас нет оснований отклонить гипотезу о том, что **средние пользовательские рейтинги платформ Xbox One и PC равны**. Таким образом, **наше предположение о том, что средние рейтинги на этих двух платформах одинаковые, можно признать обоснованным**.


* Также мы статистически обосновали наше предположение о том, что **средние пользовательские рейтинги жанров Action и Sports различаются**, и мы можем утверждать, что данное различие является статистически значимым.

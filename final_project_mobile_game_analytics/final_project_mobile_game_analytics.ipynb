# Выпускной проект: Аналитика мобильной игры на основе паттернов поведения пользователей

## Описание проекта:

В данном исследовании проводится комплексный анализ поведения пользователей мобильной игры для последующего формирования эффективной модели монетизации. В рамках анализа рассматривается одна когорта игроков, пришедшая в игру с 4 по 10 мая 2020 включительно. Исследуемый период активности игроков: c 2020-05-04 по 2020-06-05.

**Цели исследования:**

* Проанализировать распределение различных событий по всем игрокам
* Разделить игроков на группы исследователей и воинов в зависимости от выбранной стратегии прохождения уровня
* Выявить характерные различия между исследователями и воинами
* Сформулировать и проверить гипотезы статистическими методами
* Проанализировать эффективность привлечения игровой аудитории на основе расчета CAC по рекламным источникам
* Дать рекомендации по дальнейшей монетизации игры

## Главные выводы:

### 1. Показатели по игрокам
* Всего игроков в исследуемой выборке: 13576, из них 5817 игроков (42.85%) завершили 1-ый уровень, а оставшиеся 7759 (57.15%) - не завершили.
* Из 5817 игроков, завершивших 1-ый уровень, к группе воинов (прошли уровень через победу над врагом) можно отнести 3951 игроков (67.92%), а к группе исследователей (победили с помощью освоения технологии орбитальной сборки спутников) - 1866 игроков (32.08%).
* Медианное число сборочных цехов по всем игрокам – 4: при этом среди воинов медиана также равна 4, а среди исследователей - 5.
* Медианное число космопортов по всем игрокам – 4; среди воинов – 4; среди исследователей – 6.

* Половина игроков успели построить первый исследовательский центр менее чем за 206 часов с первого посещения.
* Время необходимое для прохождения 1-го уровня по всем игрокам: среднее время - 284 часа, медианное время - 277 часов. Минимальное время прохождения - 9 часов, максимальное - 741 час. Для воинов среднее время – 266 часов, медиана 259 часов; для исследователей среднее время 322.5 часа, медиана – 312 часов.

### 2. Характерные различия между игровым поведением воинов и исследователей
* Воины менее охотно прибегают к строительству и строят только в случае, если дальнейшее прохождение игры невозможно. Исследователи, в свою очередь, строят наиболее активно, поскольку, это вероятно дает им преимущество в освоении новых технологий. Так, исследователи, в среднем, строят существенно больше сооружений, чем воины, по всем доступным видам.
* Некоторые воины прошли уровень без единого космопорта - возможно им удалось победить врага до того, как им вообще понадобилось строить космопорт.
* Исследователи, ожидаемо, строят исследовательские центры более активно, поскольку это дает им возможность продвинуться дальше в игре в соответствии с их стратегией. Многие воины, также ожидаемо, пренебрегают строительством исследовательских центров вовсе, поскольку концентрируются на сражениях, куда предположительно и распределяют свои основные ресурсы. При этом, и среди воинов, и среди исследователей дальнейшее увеличение числа исследовательских центров не популярно - возможно, рост числа центров после определенной точки не приводит к дальнейшему прогрессу. Возможно, более выгодной стратегией для игроков является улучшение и апгрейд уже имеющихся исследовательских центров, а не строительство новых.
* Помимо более активного строительства, исследователи также быстрее приступают к постройкам новых типов строений (как для космопорта, так и для исследовательского центра). Возможно, это обусловлено технологическим превосходством исследователей над воинами.
* Воины значительно быстрее проходят первый уровень, по сравнению с исследователями. Это может объясняться тем, что в случае воинов, успешное быстрое сражение с врагом в самом начале игры является достаточным условием для перехода на следующий уровень. Исследователи же избегают конфронтации с врагами и основные ресурсы выделяют на активное строительство и исследование новых технологий, что является значительно более долгой и ресурсозатратной игровой стратегией.

### 3. Анализ рекламных источников
* Всего за исследуемый период, больше всего средств было потрачено на Яндекс.Директ (29.37% общих расходов). На втором месте Instagram - на него приходится 28.43% всех затрат на маркетинг. На третьем месте - Facebook (28.16%). Меньше всего было потрачено на Youtube - всего 14.05% от общих маркетинговых расходов.
* В период с 3 по 9 мая расходы на маркетинг неуклонно снижались по всем рекламным источникам. Количество привлеченных в игру пользователей также снижалось по всем источникам, как и рекламные затраты, рассмотренные ранее.
* Самым эффективным с точки зрения затрат (CAC) среди всех источников можно считать Youtube - 0.25 в среднем за пользователя. Чуть более дорогим, но все же довольно близким к Youtube, является Яндекс.Директ - 0.27 в среднем за пользователя. Ощутимо более дорогими являются Instagram (в среднем CAC 0.39) и Facebook (в среднем CAC 0.47).
* Стоит отметить, что несмотря на свою максимальную эффективность в отношении привлечения пользователей среди рассмотренных источников, Youtube незаслуженно занимает наименьшую долю общих рекламных расходов, что не является оптимальной аллокацией маркетинговых средств. Следует пересмотреть распределение рекламного бюджета в сторону значительного уменьшения доли Facebook и Instagram, и увеличения доли Youtube за их счет.

### 4. Проверка статистических гипотез
* Статистически значимое различие было выявлено между средним временем прохождения первого уровня среди группы воинов и среди группы исследователей - воины в среднем завершают уровень быстрее исследователей.
* В среднем исследователи строят больше сооружений, чем воины, и это различие статистически значимо.
* Не удалось найти значимых различий ни в какой из пар рекламных источников по среднему времени прохождения первого уровня.
* По количеству построек, игроки из Youtube строят в среднем незначительно меньше, чем игроки из Facebook и это различие статистически значимо. Во всех остальных парах рекламных источников не было выявлено значимых различий по числу строений.

### 5. Итоговые рекомендации в отношении будущей монетизации
* Для успешной монетизации игры по модели F2P через показ рекламы на экране строительства, следует максимизировать долю исследователей среди игроков.
* Исследователи могут быть более ценной аудиторией для игры по нескольким причинам: они строят больше зданий, чем воины, что подтверждается стат.тестами - соответственно, увидят больше рекламы на экране строительства; они раньше переходят в экран строительства при разблокировке новых типов построек; прохождение игры также занимает у исследователей больше времени (различие с воинами статистически значимо) - поэтому, они дольше пробудут в игре и опять же увидят больше рекламы; кроме того, исследователи могут быть более заинтересованы во внутриигровых покупках, таких как бустеры для ускорения времени строительства или платный апгрейд / разблокировка различных типов сооружений.
* Увеличение доли исследователей среди игровой аудитории можно попытаться достичь путем изменения игрового баланса в сторону поощрения строительства и исследования новых технологий. Кроме того, может быть эффективным применение моделей машинного обучения, которые будут на раннем этапе категоризировать игроков и выявлять потенциальных исследователей среди них.
* Следует привлекать больше игроков через Youtube в силу его наибольшей эффективности по затратам, а роль Facebook и Instagram рекомендуется уменьшить. Статистические тесты не подтверждают различия в качестве аудитории, пришедшей из различных рекламных источников ни по количеству построек (за исключением небольшого различия между Youtube и Facebook), ни по времени завершения уровня, соответственно стоит фокусироваться, на том канале, который предлагает максимальное число пользователей за определенный бюджет, то есть Youtube. В дальнейшем после внедрения монетизации с внутриигровыми покупками будет необходимо оценить LTV и ROMI по различным рекламным источникам, чтобы определить, откуда приходит наиболее платежеспособная аудитория. На данный момент, рациональнее всего просто максимизировать узнаваемость игры через наиболее эффективный по затратам канал - Youtube, а в дальнейшем возможно пересмотреть аллокацию рекламного бюджета.

#### 3.2.2 Присвоение игрокам статуса "исследователь / воин" в зависимости от выбранной стратегии для прохождения уровня (для завершивших уровень)<a name="subparagraph3_2_2"></a>

Определить стратегию игрока можно наиболее явно лишь в момент завершения уровня. Если игрок прошел уровень путем освоения технологии орбитальной сборки спутников, то мы будем считать его исследователем; а если же уровень был пройден без освоения этой технологии, значит имела место победа над первым врагом - тогда будем считать игрока воином.

Сохраним список идентификаторов игроков `users_id_lvl_1`, которые прошли уровень, чтобы далее проанализировать их поведение.

```python
users_id_lvl_1 = lvl_finished['user_id'].unique()
```

Отфильтруем логи, оставив только тех игроков, которые прошли первый уровень - `users_lvl_1`.

```python
users_lvl_1 = game_actions.query('user_id in @users_id_lvl_1')
```

Получим ID исследователей `researchers_id`.

```python
researchers_id = users_lvl_1.query('event == "project"')['user_id'].unique()
```

Аналогично получим ID воинов `warriors_id`.

```python
warriors_id = users_lvl_1.query('user_id not in @researchers_id')['user_id'].unique()
```

```python
print(f'Число исследователей среди завершивших 1 уровень: {len(researchers_id)}')
print(f'Число воинов среди завершивших 1 уровень: {len(warriors_id)}')
print(f'\nВсего игроков, завершивших 1 уровень: {n_users_lvl_1}\n')
print(f'Доля исследователей среди завершивших 1 уровень: {(len(researchers_id) / n_users_lvl_1):.2%}')
print(f'Доля воинов среди завершивших 1 уровень: {(len(warriors_id) / n_users_lvl_1):.2%}')
```

**Вывод:** очевидно, что наиболее популярной страгегией прохождения первого уровня является победа над врагом, а не освоение технологии орбитальной сборки спутников - 67.92% предпочли именно битву с врагом для прохождения этого этапа, и только 32.08% игроков завершили уровень мирным путем, как исследователи.

#### 3.2.4 Время первой постройки каждого типа зданий<a name="subparagraph3_2_4"></a>

Аналогично найдем время первого события постройки каждого типа здания и присоединим к общему датасету.

Первый сборочный цех - `first_assembly_shop`

```python
first_assembly_shop = game_actions.query('building_type == "assembly_shop"').groupby('user_id')['event_datetime'].min()
first_assembly_shop.name = 'first_assembly_shop'
game_actions = game_actions.merge(first_assembly_shop, on='user_id', how='left')
```

Первый космопорт - `first_spaceport`

```python
first_spaceport = game_actions.query('building_type == "spaceport"').groupby('user_id')['event_datetime'].min()
first_spaceport.name = 'first_spaceport'
game_actions = game_actions.merge(first_spaceport, on='user_id', how='left')
```

Первый исследовательский центр - `first_research_center`

```python
first_research_center = game_actions.query('building_type == "research_center"').groupby('user_id')['event_datetime'].min()
first_research_center.name = 'first_research_center'
game_actions = game_actions.merge(first_research_center, on='user_id', how='left')
```

#### 3.2.5 Количество времени с первого посещения до первой постройки каждого из зданий (в часах)<a name="subparagraph3_2_5"></a>

Найдем разницу во времени с первого посещения до первой постройки каждого типа здания и выразим полученные значения в часах.

Для сборочного цеха - столбец `assembly_shop_timedelta` таблицы `game_actions`.

```python
game_actions['assembly_shop_timedelta'] = (game_actions['first_assembly_shop'] - game_actions['first_visit']).astype('timedelta64[h]')
```

```python
game_actions['assembly_shop_timedelta'].value_counts()
```

Очевидно, что для сборочного цеха разница по времени между первым событием и первой постройкой всегда равна нулю, из чего мы можем заключить, что сборочный цех является первым доступным игроку зданием, которое обязательно должно быть построено за первую игровую сессию. Следовательно, далее мы не будем рассматривать данный показатель. 

Для космопорта - столбец `spaceport_timedelta` таблицы `game_actions`.

```python
game_actions['spaceport_timedelta'] = (game_actions['first_spaceport'] - game_actions['first_visit']).astype('timedelta64[h]')
```

Для исследовательского центра - столбец `research_center_timedelta` таблицы `game_actions`.

```python
game_actions['research_center_timedelta'] = (game_actions['first_research_center'] - game_actions['first_visit']).astype('timedelta64[h]')
```

#### 3.2.6 Время завершения первого уровня (для завершивших)<a name="subparagraph3_2_6"></a>

```python
game_actions = game_actions.merge(lvl_finish_time, on='user_id', how='left')
```

#### 3.2.7 Количество времени, требуемого для прохождения первого уровня (для завершивших, в часах)<a name="subparagraph3_2_7"></a>

Получим количество времени, требуемого на прохождение уровня, и сохраним в столбец `level_finish_timedelta` таблицы `game_actions`.

```python
game_actions['level_finish_timedelta'] = (game_actions['level_finish_time'] - game_actions['first_visit']).astype('timedelta64[h]')
```

#### 3.2.8 Количество построенных зданий различных типов (сборочный цех, космопорт, исследовательский центр)<a name="subparagraph3_2_8"></a>

Рассчитаем количество построенных зданий каждого типа для каждого пользователя и добавим к общему датасету.

Сборочный цех - `n_assembly_shops`

## 4. Проверка статистических гипотез<a name="paragraph4"></a>

Для проверки гипотез напишем функцию на основе t-test для сравнения средних двух генеральных совокупностей - `significance_test`. 

**T-test** - это параметрический тест, который используется для сравнения средних двух генеральных совокупностей. В наших экспериментах мы предполагаем, что две выборки были взяты из разных генеральных совокупностей с различными средними, поэтому нам подходит вариант t-test для независимых (несвязанных) выборок. Кроме того, нас интересует статистическая значимость различий в общем, а не проверка значимости того, что среднее одной группы больше или меньше другой - поэтому используется двусторонний (two-tailed) тест. Следовательно, мы будем использовать двусторонний t-test для независимых выборок.

Одной из предпосылок использования t-test является гомогенность дисперсий двух сравниваемых групп. Для проверки этого предположения используется тест Левена. 

*Нулевая гипотеза теста Левена:* дисперсии двух генеральных совокупностей гомогенны (не различаются). 

*Альтернативная гипотеза теста Левена:* дисперсии двух генеральных совокупностей различаются. 

После установления факта гомогенности дисперсии можно использовать обычный t-test, а при выявлении различий в дисперсии - будем применять модифицированный t-test для выборок с различными дисперсиями.

**Гипотезы в общем случае для всех тестов будут сформулированы следующим образом:**

*H0: нет различий между средними двух генеральных совокупностей*

*H1: различия между средними существуют*

```python
def significance_test(sample_1, sample_2, alpha):
    
    print('Среднее первой группы:', sample_1.mean())
    print('Среднее второй группы:', sample_2.mean())
    
    alpha = alpha
    
    levene = st.levene(sample_1, sample_2)
    
    if levene.pvalue < 0.05:
        print('Тест Левена: дисперсии различаются.')
        results = st.ttest_ind(sample_1, sample_2, equal_var = False)
        print('p-значение для t-test:', results.pvalue)

        if results.pvalue < alpha:
            print('Принимается альтернативная гипотеза о различии средних')
            return results.pvalue
        print('Нет оснований отвергать нулевую гипотезу о равенстве средних')
        return results.pvalue

    else:
        print('Тест Левена: дисперсии не различаются.')
        results = st.ttest_ind(sample_1, sample_2, equal_var = True)
        print('p-значение для t-test:', results.pvalue)
        
        if results.pvalue < alpha:
            print('Принимается альтернативная гипотеза о различии средних')
            return results.pvalue
        print('Нет оснований отвергать нулевую гипотезу о равенстве средних')
        return results.pvalue
```

### 4.1 Среднее время для завершения первого уровня различается между группами воинов и исследователей<a name="subparagraph4_1"></a>

## 5. Базовая модель монетизации<a name="paragraph5"></a>

Создатели мобильной игры планируют начать монетизацию по модели F2P с показом рекламы на экране строительства. Рассмотрим упрощенную гипотетическую модель юнит-экономики игры, которая принимает во внимание лишь маркетинговые расходы. Данная модель дает лишь общее представление о возможных результатах монетизации и может быть дополнена прочими метриками и факторами по мере необходимости.

В нашей модели юнитом является игрок. *Расходы:* стоимость привлечения игровой аудитории (ранее мы расчитали CAC). *Доходы:* размещение рекламы на экране строительства (за основную метрику возьмем RPM - доход от тысячи показов) - RPM может быть более безопасной метрикой на начальном этапе, поскольку мы будем получать доход от показов вне зависимости от кликов по рекламе

В качестве примерной оценки затрат на привлечение возьмем среднее значение CAC по всем источникам, расчитанное ранее. Таже возьмем значения RPM (CPM) из диапазона для баннеров на Android по [ссылке](https://monetizepros.com/cpm-rate-guide/mobile/) - от 0.15 до 1.5. Будем считать что, в среднем все пользователи заходят на экран постройки около 9 раз, чтобы построить здания, где и увидят рекламу. Разумеется, в реальности, игроки будут заходить на экран строительства гораздо чаще - либо чтобы проверить время до конца стройки, либо, чтобы спланировать дальнейшие действия. Поэтому предположим, что на каждое здание игрок будет заходить на экран постройки 15 раз (для установления точного количества необходимо больше данных).

```python
def unit_economics(units, rpm):    
    costs = units * cac_pivot.mean().mean()
    revenue = units * (rpm / 1000) * n_buildings_total_by_users.mean() * 15
    return revenue - costs
```

Пессимистичный сценарий: 1000 показов стоит $0.15:

```python
print(f'Прибыль/убыток на 1 игрока: {unit_economics(1, 0.15)}')
```

Нейтральный сценарий: 1000 показов стоит $0.825 (среднее значение диапазона)

```python
print(f'Прибыль/убыток на 1 игрока: {unit_economics(1, 0.825)}')
```

Оптимистичный сценарий: 1000 показов стоит $1.5

```python
print(f'Прибыль/убыток на 1 игрока: {unit_economics(1, 1.5)}')
```

Попробуем увеличить доход за счет показа видео вместо баннера. Средний доход за 1000 просмотров возьмем равным 3 ([Источник](https://www.businessofapps.com/ads/research/mobile-app-advertising-cpm-rates/))

Как мы можем увидеть, размещение баннеров в диапазоне от 0.15 до 1.5 долларов приводит к отрицательной юнит-экономике, и дальнейшее привлечение пользователей только увеличит убыток. Если же мы рассмотрим показ видео вместо баннера, то получим положительную юнит-экономику и при увеличении игровой аудитории до миллиона пользователей, можем рассчитывать на доход в более чем 80 тысяч долларов. В качестве альтернативного/дополнительного пути монетизации можно рассмотреть внутриигровые покупки, которые не будут раздражать игроков. Кроме того, чтобы не оттолкнуть игроков, рекомендуется начать показ рекламы с небольших баннеров, а более тяжелую рекламу запускать по времени ближе к половине завершения уровня - начиная со 142 часов с первой игровой сессии (основываясь на том, что в среднем у всех игроков проходит 284 часа с первой сессии до конца первого уровня). ### 6.3 Характерные различия между игровым поведением воинов и исследователей<a name="subparagraph6_3"></a>

* Воины менее охотно прибегают к строительству и строят только в случае, если дальнейшее прохождение игры невозможно. Исследователи, в свою очередь, строят наиболее активно, поскольку, это вероятно дает им преимущество в освоении новых технологий. Так, исследователи, в среднем, строят существенно больше сооружений, чем воины, по всем доступным видам.


* Некоторые воины прошли уровень без единого космопорта - возможно им удалось победить врага до того, как им вообще понадобилось строить космопорт.


* Исследователи, ожидаемо, строят исследовательские центры более активно, поскольку это дает им возможность продвинуться дальше в игре в соответствии с их стратегией. Многие воины, также ожидаемо, пренебрегают строительством исследовательских центров вовсе, поскольку концентрируются на сражениях, куда предположительно и распределяют свои основные ресурсы. При этом, и среди воинов, и среди исследователей дальнейшее увеличение числа исследовательских центров не популярно - возможно, рост числа центров после определенной точки не приводит к дальнейшему прогрессу. Возможно, более выгодной стратегией для игроков является улучшение и апгрейд уже имеющихся исследовательских центров, а не строительство новых.


* Помимо более активного строительства, исследователи также быстрее приступают к постройкам новых типов строений (как для космопорта, так и для исследовательского центра). Возможно, это обусловлено технологическим превосходством исследователей над воинами.


* Воины значительно быстрее проходят первый уровень, по сравнению с исследователями. Это может объясняться тем, что в случае воинов, успешное быстрое сражение с врагом в самом начале игры является достаточным условием для перехода на следующий уровень. Исследователи же избегают конфронтации с врагами и основные ресурсы выделяют на активное строительство и исследование новых технологий, что является значительно более долгой и ресурсозатратной игровой стратегией.

### 6.4 Анализ рекламных источников<a name="subparagraph6_4"></a>

* Всего за исследуемый период, больше всего средств было потрачено на Яндекс.Директ (29.37% общих расходов). На втором месте Instagram - на него приходится 28.43% всех затрат на маркетинг. На третьем месте - Facebook (28.16%). Меньше всего было потрачено на Youtube - всего 14.05% от общих маркетинговых расходов.


* В период с 3 по 9 мая расходы на маркетинг неуклонно снижались по всем рекламным источникам. Количество привлеченных в игру пользователей также снижалось по всем источникам, как и рекламные затраты, рассмотренные ранее.


* Самым эффективным с точки зрения затрат (CAC) среди всех источников можно считать Youtube - 0.25 в среднем за пользователя. Чуть более дорогим, но все же довольно близким к Youtube, является Яндекс.Директ - 0.27 в среднем за пользователя. Ощутимо более дорогими являются Instagram (в среднем CAC 0.39) и Facebook (в среднем CAC 0.47).


* Стоит отметить, что несмотря на свою максимальную эффективность в отношении привлечения пользователей среди рассмотренных источников, Youtube незаслуженно занимает наименьшую долю общих рекламных расходов, что не является оптимальной аллокацией маркетинговых средств. *Следует пересмотреть распределение рекламного бюджета в сторону значительного уменьшения доли Facebook и Instagram, и увеличения доли Youtube за их счет.*

### 6.5 Проверка статистических гипотез<a name="subparagraph6_5"></a>

* Статистически значимое различие было выявлено между средним временем прохождения первого уровня среди группы воинов и среди группы исследователей - *воины в среднем завершают уровень быстрее исследователей*.


* *В среднем исследователи строят больше сооружений, чем воины*, и это различие статистически значимо.


* *Не удалось найти значимых различий* ни в какой из пар рекламных источников по среднему времени прохождения первого уровня.


* По количеству построек, *игроки из Youtube строят в среднем незначительно меньше, чем игроки из Facebook* и это различие статистически значимо. *Во всех остальных парах рекламных источников не было выявлено значимых различий по числу строений.*

### 6.6 Итоговые рекомендации в отношении будущей монетизации<a name="subparagraph6_6"></a>

* Для успешной монетизации игры по модели F2P через показ рекламы на экране строительства, *следует максимизировать долю исследователей среди игроков*.


* *Исследователи могут быть более ценной аудиторией* для игры по нескольким причинам: *они строят больше зданий*, чем воины, что подтверждается стат.тестами - соответственно, увидят больше рекламы на экране строительства; *они раньше переходят в экран строительства* при разблокировке новых типов построек; *прохождение игры также занимает у исследователей больше времени* (различие с воинами статистически значимо) - поэтому, они дольше пробудут в игре и опять же увидят больше рекламы; кроме того, исследователи *могут быть более заинтересованы во внутриигровых покупках*, таких как бустеры для ускорения времени строительства или платный апгрейд / разблокировка различных типов сооружений.


* Увеличение доли исследователей среди игровой аудитории можно попытаться достичь путем изменения игрового баланса в сторону поощрения строительства и исследования новых технологий. Кроме того, может быть эффективным применение моделей машинного обучения, которые будут на раннем этапе категоризировать игроков и выявлять потенциальных исследователей среди них.


* *Следует привлекать больше игроков через Youtube* в силу его наибольшей эффективности по затратам, а роль Facebook и Instagram рекомендуется уменьшить. Статистические тесты не подтверждают различия в качестве аудитории, пришедшей из различных рекламных источников ни по количеству построек, ни по времени завершения уровня, соответственно стоит фокусироваться, на том канале, который предлагает максимальное число пользователей за определенный бюджет, то есть Youtube. В дальнейшем после внедрения монетизации с внутриигровыми покупками будет необходимо оценить LTV и ROMI по различным рекламным источникам, чтобы определить, откуда приходит наиболее платежеспособная аудитория. На данный момент, рациональнее всего просто максимизировать узнаваемость игры через наиболее эффективный по затратам канал - Youtube, а в дальнейшем возможно пересмотреть аллокацию рекламного бюджета.

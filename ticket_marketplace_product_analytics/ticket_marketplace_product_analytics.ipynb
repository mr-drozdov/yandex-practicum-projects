Таким образом, мы можем заключить, что пользователи не только чаще посещают сервис со стационарного компьютера, но и сами сессии длятся в среднем дольше, чем на мобильных устройствах. Рекомендуется в первую очередь уделять внимание пользователям, которые посещают сервис со стационарных устройств, но при этом также важно продолжать улучшать мобильный пользовательский опыт.

Наиболее лояльными оказались когорты июня 2017, августа 2017, сентября 2017, октября 2017 и ноября 2017.

* Минимальное время до совершения заказа, а так же самое часто встречающееся значение: 1 минута. В 50% случаев, время от первого посещения до заказа не превышает 19 минут.
* Продажи цикличны, ситуация с заказами развивается положительно и есть тенденция к увеличению.
* Связь между числом уникальных пользователей и числом заказов в день очень сильная и положительная (0.894). Согласно простой однофакторной модели линейной регрессии: 1 привлеченный посетитель конвертируется в 0.186 заказа. Это означает, нам нужно привлечь 6 посетителей (точнее 5.376) чтобы получить 1 заказ.
* Ежемесячные продажи росли до декабря 2017, где они достигли пика в 6218 заказов. После этого произошло падение до 4721 заказов в январе 2018. Затем продажи несколько восстановились, но в апреле случилось еще одно серьезное падение продаж до 3273 заказа.

* Более высокий средний чек мы можем наблюдать в когорте июня 2017, сентября 2017, декабря 2017, марта 2018.

* Связь между расходами на маркетинг и числом заказов также положительная и сильная (0.749), однако она существенно слабее, чем связь между маркетингом и DAU.

* ROMI: из всех рекламных источников за исследуемый период окупились только №1 (146.58%), №2 (110.42%), №9 (104.16%) и №5 (103.08%). Не окупились вложения в №4 (91.04%), №10 (79.21%) и №3 (37.55%).

**Итог:** Сейчас на менее рентабельные рекламные источники тратится гораздо больше средств, чем на те, что имеют намного лучшую окупаемость. Следовательно, текущее распределение маркетинговых расходов не оптимально и нуждается в существенном пересмотре. Следует сместить фокус на более рентабельные источники (№1, №2, №9, №5), а расходы на остальные - снизить и скорректировать в соответствии потребностями бизнеса.

### 2.1 Число уникальных пользователей (DAU, WAU, MAU) <a name="subparagraph2_1"></a>

Из столбца `visits['start_ts']` выделим отдельно дату (`visit_date`), номер недели (`visit_week`), номер месяца (`visit_month`) и год (`visit_year`).

```python
visits['visit_date'] = visits['start_ts'].dt.date
visits['visit_week']  = visits['start_ts'].dt.week
visits['visit_month'] = visits['start_ts'].dt.month
visits['visit_year']  = visits['start_ts'].dt.year
```

Далее рассмотрим все посещения сервиса в разрезе устройств `visits_by_device`.

```python
visits_by_device = pd.concat([visits['device'].value_counts(), visits['device'].value_counts(normalize=True)], axis=1)
visits_by_device.columns = ['number_of_visits', 'total_share']
visits_by_device
```

Как можно увидеть из таблицы выше, около **73%** всех посещений пришлось на стационарные компьютеры (`desktop`), а доля визитов с мобильных устройств (`touch`) составила около **27%**.

* Далее рассчитаем `DAU` - число уникальных пользователей в день.

Сгруппированные данные по каждому дню сохраним в `dau` (в целом), `dau_desktop` (для компьютеров), `dau_touch` (для мобильных устройств). Средние значения за исследуемый период сохраним в `dau_average`, `dau_desktop_average`, `dau_touch_average` соответственно.

```python
dau = visits.groupby('visit_date')['uid'].nunique()
dau_desktop = visits.query('device == "desktop"').groupby('visit_date')['uid'].nunique()
dau_touch = visits.query('device == "touch"').groupby('visit_date')['uid'].nunique()

dau_average = round(dau.mean())
dau_desktop_average = round(dau_desktop.mean())
dau_touch_average = round(dau_touch.mean())

print(f'Среднее количество уникальных пользователей в день (DAU): {dau_average}')
print(f'Среднее количество уникальных пользователей на стационарных устройствах в день (DAU): {dau_desktop_average}')
print(f'Среднее количество уникальных пользователей на мобильных устройствах в день (DAU): {dau_touch_average}')
```

Визуализируем, как DAU менялся с течением времени по устройствам.

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=dau.reset_index()['visit_date'], y=dau.reset_index()['uid'],
                    mode='lines',
                    name='Общее число пользователей'))
fig.add_trace(go.Scatter(x=dau_desktop.reset_index()['visit_date'], y=dau_desktop.reset_index()['uid'],
                    mode='lines',
                    name='Cтационарные устройства'))
fig.add_trace(go.Scatter(x=dau_touch.reset_index()['visit_date'], y=dau_touch.reset_index()['uid'],
                    mode='lines', name='Мобильные устройства'))
fig.update_layout(title="Число уникальных пользователей в день (DAU)",
    xaxis_title="Дата",
    yaxis_title="Количество уникальных пользователей в день",
    legend=dict(
    yanchor="top",
    y=0.99,
    xanchor="right",
    x=0.99),
    autosize=False,
    width=1000,
    height=650)
fig.show();
```

**Вывод:** на графике **DAU** мы можем наблюдать ярко выраженную цикличность. За исследуемый период показатель вырос в 3.3 раза - с 605 уникальных пользователей на 1 июня 2017 г. (455 для стационарных и 153 для мобильных устройств) до 1997 пользователей на 31 мая 2018 г. (1361 для стационарных и 655 для мобильных). 

При этом стоит отметить, что в дневных данных присутствует значительный шум и аномалии. Так, выделим 17 июля 2017 (значительный пик), 24 ноября 2017 (аномально высокий пик), 31 марта 2018 (аномальное падение), 31 мая 2018 (аномальный пик). Наиболее серьезными и важными аномалиями для более детального анализа можно считать пик *24 ноября 2017* (почти двукратный рост числа уникальных пользователей) и падение *31 марта 2018* (до 1 уникального пользователя). 

Можно предположить, что аномальный рост 24 ноября был связан с какой-либо маркетинговой акцией по привлечению внимания к сервису, которая имела большой успех. Аномальное падение 31 марта, скорее всего, произошло по причине временной недоступности сервиса или технической неисправности - после устранения проблемы, DAU вскоре вернулось к нормальным значениям.

На некоторых участках графика можно заметить, что DAU на разных устройствах имеет обратную зависимость - когда пользователи чаще заходят со стационарных устройств, они закономерно заходят реже с мобильных, и наоборот. Однако эта тенденция не всегда проявляется на всем протяжении графика. 

Постоянным можно считать то, что пользователи намного чаще предпочитают посещать сервис со стационарных компьютеров, а не с мобильных устройств.

* Аналогично рассчитаем `WAU` - число уникальных пользователей в неделю.

```python
wau = visits.groupby(['visit_year', 'visit_week'])['uid'].nunique()
wau_desktop = visits.query('device == "desktop"').groupby(['visit_year', 'visit_week'])['uid'].nunique()
wau_touch = visits.query('device == "touch"').groupby(['visit_year', 'visit_week'])['uid'].nunique()

wau_average = round(wau.mean())
wau_desktop_average = round(wau_desktop.mean())
wau_touch_average = round(wau_touch.mean())

print(f'Среднее количество уникальных пользователей в неделю (WAU): {wau_average}')
print(f'Среднее количество уникальных пользователей на стационарных устройствах в неделю (WAU): {wau_desktop_average}')
print(f'Среднее количество уникальных пользователей на мобильных устройствах в неделю (WAU): {wau_touch_average}')
```

```python
plt.figure(figsize=(16,8))
wau.plot(legend=True, label='Общее число пользователей')
wau_desktop.plot(legend=True, label='Cтационарные устройства')
wau_touch.plot(legend=True, label='Мобильные устройства')
plt.title('Число уникальных пользователей в неделю (WAU)')
plt.xlabel('Дата: (год, неделя)')
plt.ylabel('Количество уникальных пользователей в неделю')
plt.grid(True, alpha=0.3);
```

**Вывод:** на графике **WAU**, данные более сглажены, и выбросы не так заметны. Здесь мы более отчетливо видим общие тренды - до 47 недели 2017 года число уникальных пользователей росло, а после - начало стабильно снижаться.

```python
mau = visits.groupby(['visit_year', 'visit_month'])['uid'].nunique()
mau = visits.groupby(['visit_year', 'visit_month'])['uid'].nunique()
mau_desktop = visits.query('device == "desktop"').groupby(['visit_year', 'visit_month'])['uid'].nunique()
mau_touch = visits.query('device == "touch"').groupby(['visit_year', 'visit_month'])['uid'].nunique()

mau_average = round(mau.mean())
mau_desktop_average = round(mau_desktop.mean())
mau_touch_average = round(mau_touch.mean())

print(f'Среднее количество уникальных пользователей в месяц (MAU): {mau_average}')
print(f'Среднее количество уникальных пользователей на стационарных устройствах в месяц (MAU): {mau_desktop_average}')
print(f'Среднее количество уникальных пользователей на мобильных устройствах в месяц (MAU): {mau_touch_average}')
```

```python
plt.figure(figsize=(16,8))
mau.plot(legend=True, label='Общее число пользователей')
mau_desktop.plot(legend=True, label='Cтационарные устройства')
mau_touch.plot(legend=True, label='Мобильные устройства')
plt.title('Число уникальных пользователей в месяц (MAU)')
plt.xlabel('Дата: (год, месяц)')
plt.ylabel('Количество уникальных пользователей в месяц')
plt.grid(True, alpha=0.3);
```

**Вывод:** график **MAU**, еще больше обобщает данные и демонстрирует нам ту же картину, что и WAU - ноябрь 2017 года был переломной точкой, после которой число уникальных пользователей начало снижаться. Требуется изучить, что именно привело к стабильному снижению интереса к продукту. Можем предположить, что такая динамика метрики укладывается в естественную сезонность в индустрии мероприятий, но эта гипотеза нуждается в проверке на более длинном временном периоде.

* Далее рассчитаем Sticky Factor. Эта метрика показывает степень лояльности и вовлеченности аудитории

```python
dau_by_wau = dau_average / wau_average
wau_by_mau = wau_average / mau_average
print(f'Sticky factor DAU/WAU: {dau_by_wau:.2f}')
print(f'Sticky factor WAU/MAU: {wau_by_mau:.2f} \n')

dau_by_wau_desktop = dau_desktop_average / wau_desktop_average
wau_by_mau_desktop = wau_desktop_average / mau_desktop_average
print(f'Sticky factor DAU/WAU (desktop): {dau_by_wau_desktop:.2f}')
print(f'Sticky factor WAU/MAU (desktop): {wau_by_mau_desktop:.2f} \n')

dau_by_wau_touch = dau_touch_average / wau_touch_average
wau_by_mau_touch = wau_touch_average / mau_touch_average
print(f'Sticky factor DAU/WAU (touch): {dau_by_wau_touch:.2f}')
print(f'Sticky factor WAU/MAU (touch): {wau_by_mau_touch:.2f}')
```

**Вывод:** Sticky factor находится на хорошем уровне, что свидетельствует о достаточном уровне вовлеченности и лояльности среди пользователей продукта - они возвращаются в сервис. Также стоит отметить, что для мобильных устройств данный показатель несколько ниже, чем для десктопов. Возможно, это объясняется тем, что посетив сервис с мобильного устройства, пользователи могут вернуться уже с компьютера, когда им это более удобно. И возможно сам процесс поиска мероприятий и покупки билетов более комфортен именно на стационарном компьютере.

### 2.4 Retention Rate и Churn Rate <a name="subparagraph2_4"></a>

* Расчет Retention Rate

Найдем дату первого посещения и сохраним в `first_visit_date`.

```python
first_visit_date = visits.groupby('uid')['start_ts'].min()
first_visit_date.name = 'first_visit_date'
```

Соединим полученные данные с датасетом `visits`.

```python
visits = pd.merge(visits, first_visit_date, on='uid')
```

Приведем размерность столбцов `first_visit_month`, `visit_month` к месяцам.

```python
visits['first_visit_month'] = visits['first_visit_date'].astype('datetime64[M]')
visits['visit_month'] = visits['start_ts'].astype('datetime64[M]')
```

Рассчитаем lifetime когорт в `cohort_lifetime` как разность между датами текущего и первого посещений. Разделим на число дней в месяце и округлим результат.

```python
visits['cohort_lifetime'] = ((visits['visit_month'] - visits['first_visit_month']) / np.timedelta64(1,'M')).round().astype('int')
```

Сгруппируем данные по когортам в зависимости от месяца первого посещения - `cohorts`.

```python
cohorts = visits.groupby(['first_visit_month', 'cohort_lifetime'])['uid'].nunique().reset_index()
cohorts['first_visit_month'] = cohorts['first_visit_month'].dt.strftime('%Y-%m') 
```

Посчитаем количество пользователей в когортах на нулевой период lifetime - `initial_users_count`.

```python
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_visit_month','uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'}) 
```

Присоединим `initial_users_count` к `cohorts`.

```python
cohorts = cohorts.merge(initial_users_count,on='first_visit_month') 
```

Расчитаем `retention` как отношение количества пользователей на данный период к первоначальному количеству (на нулевой период).

```python
cohorts['retention'] = cohorts['uid'] / cohorts['cohort_users']
```

Соберем данные когортного анализа в сводную таблицу `retention_pivot`. В `retention_pivot_1` удалим нулевой период для наглядности.

```python
retention_pivot = cohorts.pivot_table(index='first_visit_month',columns='cohort_lifetime',values='retention',aggfunc='sum')
retention_pivot_1 = retention_pivot.drop(0, axis=1)
```

Визуализируем таблицу с помощью тепловой карты.

```python
plt.figure(figsize=(13, 9))
plt.title('Cohorts: User Retention')
sns.heatmap(retention_pivot_1, annot=True, fmt='.1%', linewidths=1, linecolor='gray', vmin=0.0,cmap='Reds')
plt.xlabel('Lifetime когорты')
plt.ylabel('Когорта по первому посещению'); 
```

**Вывод:** Результаты когортного анализа демонстрируют, что коэффициент удержания пользователей не высокий, но приемлемый. На следующий месяц после первого визита (1 период), сервис посетили от **4.7%** (апрельская когорта) до **8.4%** (сентябрьская когорта) от первоначального числа пользователей. В целом можно считать, что наиболее лояльными оказались когорты *июня 2017, августа 2017, сентября 2017, октября 2017 и ноября 2017*. Стоит проанализировать, какие маркетинговые стратегии были использованы в тот период и усилить их, так как они позволяли более эффективно удерживать пользователей. Очевидно, что уже начиная с декабрьской когорты 2017 какие-то факторы снижают лояльность пользователей, и коэффициент удержания более существенно снижается уже начиная с первого периода.

* Расчет Churn Rate

```python
cohorts['churn_rate'] = cohorts.groupby(['first_visit_month'])['uid'].pct_change() 
churn_pivot = cohorts.pivot_table(index='first_visit_month',columns='cohort_lifetime',values='churn_rate',aggfunc='sum')
churn_pivot_1 = churn_pivot.drop(0, axis=1)
```

```python
plt.figure(figsize=(13, 9))
plt.title('Cohorts: Churn Rate')
sns.heatmap(churn_pivot_1, annot=True, fmt='.1%', linewidths=1, linecolor='gray')
plt.xlabel('Lifetime когорты')
plt.ylabel('Когорта по первому посещению'); 
```

**Вывод:** коэффициент оттока в первый месяц после привлечения пользователя составил от -91.6% (сентябрьская когорта) до -95.9% (мартовская когорта). Отдельно необходимо отметить резкий отток пользователей по всем когортам в апреле 2018. Данный факт вызывает беспокойство и требует дальнейшего анализа - в этот период когорта февраля испытала наибольший отток: -57.1%. При этом стоит обратить внимание и на положительные моменты, когда churn rate был положительным - например в когорте июня на 3 периоде (то есть в сентябре 2017) пользователи возвращались, их число выросло на 17.2% по сравнению с предыдущим периодом. Также положительные сдвиги наблюдались в этой когорте и в нескольких будущих периодах. Это говорит о том, что степень оттока в июньской когорте была меньше чем, среди других когорт. Возможно пользователям этой когорты были предложены особые условия, которые повлияли на снижение оттока. Также положительными можно назвать результаты когорты июля 2017 г.

### 3.4 LTV <a name="subparagraph3_4"></a>

В данном проекте мы будем использовать упрощенное понимание LTV и принимать маржинальность сервиса за 100%. Для более детального анализа формула может быть изменена при получении дополнительных данных.

```python
ltv_cohort_sizes = orders.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
ltv_cohort_sizes.columns = ['first_order_month', 'n_buyers']
```

```python
ltv_cohorts = orders.groupby(['first_order_month','order_month']).agg({'revenue': 'sum'}).reset_index()
ltv_report = pd.merge(ltv_cohort_sizes, ltv_cohorts, on='first_order_month')
```

```python
ltv_report['age'] = (ltv_report['order_month'] - ltv_report['first_order_month']) / np.timedelta64(1, 'M')
ltv_report['age'] = ltv_report['age'].round().astype('int')
ltv_report['ltv'] = ltv_report['revenue'] / ltv_report['n_buyers']
ltv_output = ltv_report.pivot_table(
        index='first_order_month', 
        columns='age', 
        values='ltv', 
        aggfunc='mean')
ltv_output.round(3).fillna('')
```

```python
print('Суммарное LTV по когортам:')
for cohort in ltv_output.index:
    print(f'{cohort.strftime("%Y-%m")} |', ltv_output.loc[cohort].sum().round(2))
```

**Вывод:** за исследуемый период когорта сентября 2017 принесла сервису наибольшую ценность, за ней следует когорта июня 2017.

### 4.2 CAC <a name="subparagraph4_2"></a>

Сначала найдем первый источник, по которому покупатели в первый раз перешли в сервис - `first_source_id`.

```python
first_visits = visits.query('first_visit_date == start_ts')
first_source_id = first_visits[['uid', 'source_id']].copy()
```

Добавим эту информацию к имеющимся данным.

```python
ltv_calc = pd.merge(orders, first_source_id, on='uid', how='inner')
```

Далее найдем общую выручку по каждому рекламному источнику `total_revenue_by_sources`, а также количество клиентов, пришедших из соответствующих источников `n_customers_by_sources`.

```python
total_revenue_by_sources = ltv_calc.groupby('source_id')['revenue'].sum().reset_index()
n_customers_by_sources = ltv_calc.groupby('source_id')['uid'].nunique().reset_index()
```

Соберем всю необходимую информацию в датафрейме `report`.

```python
report = total_revenue_by_sources.merge(total_costs_by_sources, on='source_id', how='left')
report = report.merge(n_customers_by_sources, on='source_id', how='left')
```

Рассчитаем средний CAC как отношение общих расходов на различные источники к количеству клиентов, привлеченных через них.

```python
report['cac'] = report['costs'] / report['uid']
report[['source_id', 'cac']].sort_values(ascending=False, by='cac')
```

**Вывод:** Согласно усредненным значениям CAC, самым дорогим источником оказался **№3** (13.79 руб. на привлечение одного клиента). За ним следует **№2** (12.18 руб.) и **№5** (7.56 руб.). Самый дешевый источник **№10** (4.46 руб. на клиента). По расходам на источник №7 данные отсутствуют.

### 4.3 Рентабельность вложений в маркетинг (ROMI)  <a name="subparagraph4_3"></a>

Далее расчитаем окупаемость вложений в маркетинг `romi` как `ltv`, разделенное на `cac`.

```python
report['ltv'] = report['revenue'] / report['uid']
report['romi'] = report['ltv'] / report['cac']
report.sort_values(ascending=False, by='romi').drop(5, axis=0)
```

**Вывод:** Из всех рекламных источников за исследуемый период **окупились** только №1 (146.58%), №2 (110.42%), №9 (104.16%) и №5 (103.08%). **Не окупились** вложения в №4 (91.04%), №10 (79.21%) и №3 (37.55%).

Таким образом, текущее распределение маркетинговых расходов является неоптимальным и нуждается в изменении. Следует сместить фокус на более рентабельные источники (№1, №2, №9, №5), а расходы на остальные - снизить и скорректировать в соответствии потребностями бизнеса.

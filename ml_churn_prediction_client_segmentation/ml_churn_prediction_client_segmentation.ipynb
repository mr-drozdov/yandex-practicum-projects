# Прогнозирование оттока и сегментация клиентов фитнес-центра с помощью ML-методов

## Описание проекта:

В данном проекте проводится анализ данных по клиентам фитнес-центра с использованием методов машинного обучения. ## Главные выводы:

* **Общая информация о клиентах.** Клиенты разделены по полу примерно поровну. 84.5% клиентов проживают или работают в районе, где находится фитнес-центр. 48.7% клиентов являются сотрудниками компании-партнера, а значит имеют скидку на посещения. 30.85% клиентов пришли по промокоду в рамках акции "приведи друга". 90.35% клиентов указали номер телефона. 41.23% клиентов посещали групповые занятия. Средний и медианный возраст клиентов - 29 лет. Общая выручка от дополнительных услуг с клиента в среднем 146.94; медиана 136.22. В среднем до конца контракта у клиентов осталось 4.32 месяца, медианное значение 1 месяц. Среднее время с момента первого посещения 3.72 месяца, медиана 3 месяца. Средняя частота посещения занятий в неделю за все время 1.88 занятий, а среднее значение за предыдущий месяц 1.77 занятий. 26.53% клиентов из выборки ушли в текущем месяце.


* **Модель прогнозирования оттока клиентов.** Были рассмотрены две модели: логистическая регрессия и random forest. Обе модели показали хорошие результаты, но чуть более точной оказалась модель логистической регрессии - рекомендуется выбрать ее. В обеих моделях самыми значимыми факторами предсказания оттока являются: lifetime - чем дольше клиент посещает фитнес-центр, тем больше это входит в привычку и тем меньше вероятность того, что он перестанет в него ходить; кроме того, важную роль играет среднее посещение занятий (как в текущем месяце, так и в целом) - чем чаще клиент посещает центр, тем более он мотивирован и менее склонен бросать занятия спортом; возраст - чем клиент старше, тем менее вероятен отток - предположтельно с возрастом клиенты начинают более осознанно подходить к здоровому образу жизни и стараются следить за здоровьем, при этом в абсолютном выражении разница в возрасте не обязательно должна быть существенной, чтобы повлиять на занятия спортом; длительность абонемента / время до конца абонемента - чем более длительный абонемент приобретает клиент, а также чем больше времени остается до конца абонемента, тем менее вероятен отток - более мотивированные клиенты пользуются долгосрочными предложениями, чтобы сэкономить, зная, что продолжат заниматься спортом, а менее лояльные клиенты чаще всего приобретают абонемент на 1 месяц, чтобы попробовать, и возможно не захотят его продлевать; траты на дополнительные услуги - клиенты, которые тратят больше на доп.услуги менее склонны к оттоку, так как их, скорее всего, устраивает фитнес-центр, и они стремятся максимизировать свой комфорт за дополнительные деньги во время пребывания в нем.


* **Сегментирование клиентов.** Были выделены 5 основных кластеров (сегментов) среди клиентов (№0,№1,№2,№3,№4). А вместе кластеры №3, №2 и №1 составляют 53.79% всех клиентов - таким образом больше половины всех клиентов находятся в зоне риска по оттоку. Безопасные кластеры (с крайней низкой долей оттока) №0 и №4 вместе составляют 46.17% клиентов.Важные особенности надежных кластеров. Сотрудников компаний-партнеров больше именно в лояльных кластерах. Подавляющее большинство клиентов из лояльных кластеров имеют годовой абонемент, а нелояльные клиенты ограничиваются абонементами длительностью 1 месяц. Более лояльные сегменты относительно старше, чем менее лояльные. Кроме того, клиенты из лояльных сегментов тратят относительно больше на дополнительные услуги центра, по сравнению с менее надежными сегментами. В лояльных сегментах выше значение lifetime. Частота посещений в текущем месяце и в целом у лояльных кластеров также выше.


* **Важные особенности надежных кластеров.** Сотрудников компаний-партнеров больше именно в лояльных кластерах. Акция "приведи друга" также пользуется спросом среди лояльных сегментов. ## 4. Прогнозирование оттока клиентов<a name="paragraph4"></a>

Построим модель бинарной классификации клиентов, приняв факт оттока (`churn`) за целевую переменную.

### 4.1 Разделение датасета на обучающую и валидационную выборку<a name="subparagraph4_1"></a>

Делим данные в пропорции 80/20.

```python
y = data['churn']
x = data.drop(['churn'], axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

### 4.2 Обучение модели с помощью Logistic Regression и Random Forest<a name="subparagraph4_2"></a>

* Логистическая регрессия

Чтобы избежать проблемы мультиколлинеарности между некоторыми переменными, используем параметр `l1`, который применит L1-регуляризацию (lasso). Кроме того, будем использовать стандартизированные значения признаков в модели.

```python
scaler = StandardScaler()
x_train_sc = scaler.fit_transform(x_train)
x_test_sc = scaler.transform(x_test)
```

```python
lr = LogisticRegression(random_state=0, solver='saga', penalty='l1')
lr.fit(x_train_sc, y_train)
```

```python
lr_weights = pd.DataFrame(columns=x.columns, data=lr.coef_).T
lr_weights.columns = ['weights']
lr_weights['abs_weights'] = abs(lr_weights['weights'])
lr_weights_sort = lr_weights.sort_values(by='abs_weights', ascending=False)
lr_weights_sort
```

```python
plt.figure(figsize=(16,8))
plt.title('Абсолютный вес признаков в модели Logistic Regression')
sns.barplot(data=lr_weights_sort, y=lr_weights_sort.index, x='abs_weights')
plt.ylabel('Признак')
plt.xlabel('Абсолютный вес признака')
sns.despine()
plt.show()
```

```python
lr_prob = lr.predict_proba(x_test_sc)[:,1]
```

```python
lr_y_pred = lr.predict(x_test_sc)
```

**Вывод:** наибольшее влияние на предотвращение оттока оказывает средняя частота посещения занятий в текущем месяце (`avg_class_frequency_current_month`). Кроме того, большое влияние на удержание клиентов оказывает `lifetime` и `age` (имеют отрицательный коэффициент). Любопытно, что среднее количество занятий в целом (`avg_class_frequency_total`) имеет положительный и большой по модулю коэффициент. Это противоречит логике - большая частота посещения в целом не должна приводить к увеличению вероятности оттока клиента. Не исключено, что аномальный коэффициент возник из-за сильной корреляции между `avg_class_frequency_current_month` и `avg_class_frequency_total`, а lasso-регуляризация сработала не эффективно. Тем не менее, при удалении любого признака из обеих скоррелированных пар, все метрики ощутимо ухудшаются на валидации при несущественном изменении коэффициентов, что может свидетельствовать о том, что все эти признаки несут важную информацию, несмотря на то, что могут в некоторой степени дублировать друг друга. Вследствие этого, было принято решение не устранять пары скоррелированных признаков в модели.

* Random Forest

Для ансамблевых моделей стандартизация признаков не требуется.

```python
rf = RandomForestClassifier(random_state=0, n_estimators=100)
rf.fit(x_train, y_train)
```

```python
rf_importances = pd.DataFrame({'feature':x.columns, 'importance':rf.feature_importances_}).sort_values(by='importance', ascending=False)
rf_importances
```

```python
plt.figure(figsize=(16,8))
plt.title('Важность признаков в модели Random Forest')
sns.barplot(data=rf_importances, y='feature', x='importance')
plt.ylabel('Признак')
plt.xlabel('Важность')
sns.despine()
plt.show()
```

```python
rf_prob = rf.predict_proba(x_test)[:,1]
```

```python
rf_y_pred = rf.predict(x_test)
```

**Вывод:** в модели Random Forest признаки по своему вкладу выстроились в порядке, похожем на модель логистической регрессии, с некоторыми отличиями. Самую главную роль играет `lifetime`, затем идут `avg_class_frequency_current_month`, `avg_class_frequency_class_total`, `age`, `avg_additional_charges_total`,`month_to_end_contract`, `contract_period`. Остальные факторы не являются существенными для модели.

### 4.3 Оценка и сравнение моделей на валидационной выборке<a name="subparagraph4_3"></a>

```python
def metrics(y_test, y_pred, y_prob):
    
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mean_abs_error = mean_absolute_error(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_prob)
    
    return print(
        f''' 
        | accuracy  | {accuracy: >7.4f} |
        | precision | {precision: >7.4f} |
        | recall    | {recall: >7.4f} |
        | f1_score  | {f1: >7.4f} |
        | r2_score  | {r2: >7.4f} |
        | MAE       | {mean_abs_error: >7.4f} |
        | roc_auc   | {roc_auc: >7.4f} |
        ''')
```

```python
print('Метрики для логистической регрессии:')
metrics(y_test, lr_y_pred, lr_prob)
```

```python
print('Метрики для случайного леса:')
metrics(y_test, rf_y_pred, rf_prob)
```

**Вывод:** на основании метрик, обе модели можно считать достаточно качественными и пригодными для анализа и прогнозирования оттока клиентов. При этом, по всем метрикам логистическая регрессия в текущей спецификации превосходит модель Random Forest. Логистическая регрессия имеет более высокие показатели доли правильных ответов, точности, полноты, F1, коэффициент детерминации R2, и площади под ROC кривой, а также более низкий средний модуль ошибки. Следовательно, целесобразно использовать модель логистической регрессии для поставленной задачи.

## 6. Общий вывод и рекомендации<a name="paragraph6"></a>

### 6.1 Основые итоги исследования<a name="subparagraph6_1"></a>

- **Общая информация о клиентах.** Клиенты разделены по полу примерно поровну. 84.5% клиентов проживают или работают в районе, где находится фитнес-центр. 48.7% клиентов являются сотрудниками компании-партнера, а значит имеют скидку на посещения. 30.85% клиентов пришли по промокоду в рамках акции "приведи друга". 90.35% клиентов указали номер телефона. 41.23% клиентов посещали групповые занятия. Средний и медианный возраст клиентов - 29 лет. Общая выручка от дополнительных услуг с клиента в среднем 146.94; медиана 136.22. В среднем до конца контракта у клиентов осталось 4.32 месяца, медианное значение 1 месяц. Среднее время с момента первого посещения 3.72 месяца, медиана 3 месяца. Средняя частота посещения занятий в неделю за все время 1.88 занятий, а среднее значение за предыдущий месяц 1.77 занятий. 26.53% клиентов из выборки ушли в текущем месяце.


- **Корреляционные связи.** *Пол и наличие телефона в анкете* - не связаны ни с другими признаками, ни с целевой переменной оттока. Все остальные признаки имеют отрицательную связь с оттоком - т.е. при увеличении признаков вероятность оттока снижается.


- **Модель прогнозирования оттока клиентов.** Были рассмотрены две модели: *логистическая регрессия и random forest*. Обе модели показали хорошие результаты, но чуть более точной оказалась модель *логистической регрессии* - рекомендуется выбрать ее. В обеих моделях самыми значимыми факторами предсказания оттока являются: *lifetime* - чем дольше клиент посещает фитнес-центр, тем больше это входит в привычку и тем меньше вероятность того, что он перестанет в него ходить; кроме того, важную роль играет *среднее посещение занятий (как в текущем месяце, так и в целом)* - чем чаще клиент посещает центр, тем более он мотивирован и менее склонен бросать занятия спортом; *возраст* - чем клиент старше, тем менее вероятен отток - предположтельно с возрастом клиенты начинают более осознанно подходить к здоровому образу жизни и стараются следить за здоровьем, при этом в абсолютном выражении разница в возрасте не обязательно должна быть существенной, чтобы повлиять на занятия спортом; *длительность абонемента / время до конца абонемента* - чем более длительный абонемент приобретает клиент, а также чем больше времени остается до конца абонемента, тем менее вероятен отток - более мотивированные клиенты пользуются долгосрочными предложениями, чтобы сэкономить, зная, что продолжат заниматься спортом, а менее лояльные клиенты чаще всего приобретают абонемент на 1 месяц, чтобы попробовать, и возможно не захотят его продлевать; *траты на дополнительные услуги* - клиенты, которые тратят больше на доп.услуги менее склонны к оттоку, так как их, скорее всего, устраивает фитнес-центр, и они стремятся максимизировать свой комфорт за дополнительные деньги во время пребывания в нем. 


- **Сегментирование клиентов.** Были выделены *5 основных кластеров* (сегментов) среди клиентов (№0,№1,№2,№3,№4). *Самыми лояльными* являются кластеры №0 и №4 - в них отток составил всего 2.77% и 6.81% соответственно. Кластер №1 можно условно назвать *умеренно лояльным* - уровень оттока 26.75%. А *наименее лояльными (ненадежными)* являются кластеры №2 и №3, которые демонстрируют отток 44.36% и 51.43% соответственно. Крайне важно отметить, что *кластер №3 (с наивысшим оттоком) занимает наибольшую долю* среди всех кластеров (31.55%). А вместе кластеры №3, №2 и №1 составляют 53.79% всех клиентов - таким образом *больше половины всех клиентов находятся в зоне риска по оттоку*. *Безопасные кластеры* (с крайней низкой долей оттока) №0 и №4 вместе составляют 46.17% клиентов.


- **Важные особенности надежных кластеров.** *Сотрудников компаний-партнеров больше* именно в лояльных кластерах. *Акция "приведи друга" также пользуется спросом* среди лояльных сегментов. Кроме того, эти кластеры характеризуются более высокой долей клиентов, посещающих *групповые занятия*. Подавляющее большинство клиентов из лояльных кластеров *имеют годовой абонемент*, а нелояльные клиенты ограничиваются абонементами длительностью 1 месяц. Более *лояльные сегменты относительно старше*, чем менее лояльные. Кроме того, клиенты из лояльных сегментов *тратят относительно больше на дополнительные услуги* центра, по сравнению с менее надежными сегментами. В лояльных сегментах *выше значение lifetime*. *Частота посещений в текущем месяце и в целом* у лояльных кластеров также *выше*.

### 6.2 Рекомендации по уменьшению оттока клиентов<a name="subparagraph6_2"></a>

В первую очередь следует акцентировать внимание на том, чтобы существенно снизить долю ненадежных сегментов, которые принадлежат к группе риска. В особенности, это относится к кластеру №3, который является самым крупным и наименее лояльным, а также к кластеру №2, который, хоть и не имеет большого размера, также демонстрирует крайне высокие показатели оттока. На основании межкластерных различий можно предложить следующие стратегии:
1. **Чаще проводить акцию "Приведи друга"** и возможно обновить её условия. Друзья могут заниматься спортом вместе, что придает мотивацию и снижает отток.
2. **Продвигать групповые занятия** - скидки, особые предложения и реклама. Как и в предыдущем пункте, люди будут создавать социальные связи, и этот социальный фактор будет благотворно влиять на удержание клиентов.
3. **Расширить партнерскую сеть** и активно продвигать корпоративные абонементы для сотрудников компаний-партнеров на особых условиях. Сотрудники партнеров понимают, что не смогут найти подобное выгодное предложение на рынке, и это побуждает их оставаться в фитнес-центре.
4. **Активно продвигать долгосрочные абонементы** со скидкой, чтобы выработать привычку у клиента посещать фитнес-центр регулярно. Предлагать бесплатное продление, если клиент настаивает на месячном абонементе.
5. **Индивидуальные стимулирующие предложения для клиентов**, у которых снижается среднее количество посещенных занятий, снижаются/отсутствуют траты на дополнительные услуги, или заканчивается абонемент.
